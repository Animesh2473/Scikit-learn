```python
import pandas as pd
import numpy as np
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
#S1 divide Data and conclusions
x = heart_disease.drop("target",axis=1)
y = heart_disease["target"]
```

```python
x
```

```python
y
```

```python
#S2 select model that will anaylise pattern

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100)

clf.get_params()
```

```python
# S3 Now we will split a data into test and train

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
# Now fit data into model of selection
clf.fit(x_train,y_train);
```

```python
#S4 Now to see predection of the model
y_preds = clf.predict(x_test)
y_preds
```

```python
y
```

```python
clf.score(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
# here will will compare test and preadicted value
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

```python
np.random.seed(70)
for i in range(10,100,10):
    print(f"Traning model with {i} estimaters....")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train,y_train)
    print(f"Model accuracy at test score is {(clf.score(x_test,y_test))*100:2f}% \n")
```

```python
import pickle
pickle.dump(clf,open("random_forest_model_2.pkl","wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_2.pkl","rb"))
loaded_model.score(x_test,y_test)
```

```python
car_sales = pd.read_csv("car-sales-extended.csv")
car_sales.head()
```

```python
# split into x/y
x = car_sales.drop("Price",axis=1)
y = car_sales["Price"]
```

```python
# Now Convert the data into numbers 
# Make - yes
# Colour - yes 
# Odomeater (KM) - No
# Doors - Yes (As number of dors are fix and can vary in catagaries like 4,5,3 it is considered to be catagorical)
car_sales["Doors"].value_counts()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

categorical_features = ["Make", "Colour", "Doors"]

one_hot = OneHotEncoder()

transformer = ColumnTransformer([("one_hot",one_hot,categorical_features)],
                                remainder="passthrough")

transformed_x = transformer.fit_transform(x)

pd.DataFrame(transformed_x)
```

```python
dummies = pd.get_dummies(car_sales[["Make", "Colour", "Doors"]],dtype=float)
dummies
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(transformed_x,y,test_size=0.2)
```

```python
model.fit(x_train,y_train);
```

```python
model.score(x_test,y_test)
```
