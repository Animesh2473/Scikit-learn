```python
data.dropna(subset='Crash_Severity',inplace=True)

# Define diffrent features and transfor pipeline
catagorical_features=['Gender','Vehicle_Type','Road_Type','Alcohol_Consumption', 'Crash_Type', 'Seatbelt_Usage','Road_Surface_Condition']
catagorical_transformer = Pipeline(steps=[('imputer',SimpleImputer(strategy='constant',fill_value='missing')),
                                           ('onehot',OneHotEncoder(handle_unknown='ignore'))])
door_features =["Number_of_Lanes"]
door_transformer = Pipeline(steps=[('imputer',SimpleImputer(strategy='constant',fill_value= 4))])

numeric_featues = ["Age","Crash_Time","Vehicle_Speed","Lane_Width","Speed_Limit"]
numeric_transformer = Pipeline(steps=[('imputer',SimpleImputer(strategy='mean'))])



preprocessor = ColumnTransformer(transformers=[('cat',catagorical_transformer,catagorical_features),
                                               ('door',door_transformer,door_features),
                                               ('num',numeric_transformer,numeric_featues)])
                                   


model = Pipeline(steps = [('preprocessor',preprocessor),
                           ('model',RandomForestRegressor())])


x= data.drop("Crash_Severity",axis=1)
y = data["Crash_Severity"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)


model.fit(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
pipe_grid = {

    "preprocessor__num__imputer__strategy":['mean','median'],
    "model__n_estimators":[100,1000],
    "model__max_depth":[None,5],
    "model__max_features":["sqrt", "log2", None],
    "model__min_samples_split":[2,4]
}


rs_model = GridSearchCV(model,pipe_grid,cv=5,verbose=2)

rs_model.fit(x_train,y_train)
```
